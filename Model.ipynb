```python
#Since Data is Right skewed , we can fill missing values with median

def Fill_Missing_Median(df):
    for n,c in df.items():        
            df[n] = df[n].fillna(df[n].median())
            
Fill_Missing_Median(dress_sales)
```

```python
df_final=attribute.copy()

df_final.drop(['Rating','Recommendation','Dress_ID'],axis=1,inplace=True) #droping non_object datatype feature,we will 
                                                                 # add them up after converting objects to categories

#Change any columns of strings in a panda's dataframe to a column of categorical values. This applies the changes inplace
def Change_To_CategoricalValue(df):
    for n,c in df.items():
        if pd.api.types.is_object_dtype(c):
            df[n] = c.astype('category').cat.as_ordered()
            
Change_To_CategoricalValue(df_final)
```

```python
df_final.info()
```

```python
#Converting categorical features df to categorical codes And Dummies
def Change_To_Categoricalcodes(df):
    for n,c in df.items():
            df[n]=c.cat.codes
        
Change_To_Categoricalcodes(df_final)

df_final['Ratings']=attribute['Rating']
df_final['Target']=attribute['Recommendation']
df_final['Dress_ID']=attribute['Dress_ID']
```

```python
#Mergin the two DataFrame
Sales_df = pd.merge(df_final ,dress_sales,on = 'Dress_ID')
display_all(Sales_df.head())
```

```python
from sklearn.model_selection import train_test_split
X = Sales_df.drop(['Target','Dress_ID'],axis=1)
y = Sales_df["Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state =1)
```

```python
print('The number of samples into the Train data is {}.'.format(X_train.shape[0]))
print('The number of samples into the test data is {}.'.format(X_test.shape[0]))
```

```python
#Feature selection using Random Forest

from sklearn.ensemble import RandomForestClassifier
params={'random_state':0,'n_jobs':-1,'n_estimators':5000,'max_depth':8}

clf = RandomForestClassifier(**params)
clf = clf.fit(X, y)

# Plot features importances
imp = pd.Series(data=clf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10,12))
plt.title("Feature importance")
ax = sns.barplot(y=imp.index, x=imp.values, palette="Blues_d", orient='h')
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model=LogisticRegression(C=1.0, class_weight=None, dual=False,
                                          fit_intercept=True,
                                          intercept_scaling=1, l1_ratio=None,
                                          max_iter=400, multi_class='auto',
                                          n_jobs=None, penalty='l2',
                                          random_state=None, solver='lbfgs',
                                          tol=0.0001, verbose=0,
                                          warm_start=False)
logistic_model.fit(X_train,y_train)

accuracy = logistic_model.score(X_test,y_test)
print("Logistic Regression accuracy is :",accuracy*100)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV,cross_val_score
model=GradientBoostingClassifier()


model_parameters = { 
    'n_estimators' : [10,50,100,200,500],
    'max_depth' : [3,5,10,None],
    'min_samples_leaf' : [np.random.randint(1,10)] 
}

gscv_GBM=GridSearchCV(estimator=model,
                     param_grid=model_parameters,
                     cv=5,
                     verbose=1,
                     scoring='accuracy')

gscv_GBM.fit(X_train,y_train)
```

```python
print('The best parameter are -', gscv_GBM.best_params_)
```

```python
final_mod_GBM = GradientBoostingClassifier(**gscv_GBM.best_params_)
final_mod_GBM.fit(X_train, y_train)

from sklearn.metrics import classification_report
train_pred = final_mod_GBM.predict(X_train)
test_pred = final_mod_GBM.predict(X_test)

print('Classification report for test data is : \n',
      classification_report(y_test, test_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()

model_parameters = { 
        'n_estimators':[10, 50, 100, 200, 500],
        'max_depth': [3, 5, 10],
        'min_samples_leaf': [np.random.randint(1,10)]
                  }


gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    cv=5, 
                    verbose=1, 
                    scoring='accuracy')

gscv.fit(X_train, y_train)  
```

```python
print('The best parameter are -', gscv.best_params_)
```

```python
final_mod_xgb = XGBClassifier(max_depth= 5, min_samples_leaf= 2, n_estimators= 500,learning_rate=0.1)

final_mod_xgb.fit(X_train, y_train)

from sklearn.metrics import classification_report
test_pred = final_mod_xgb.predict(X_test)

print('Classification report for test data is : \n',
      classification_report(y_test, test_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm_lr = confusion_matrix(y_test,final_mod_xgb.predict(X_test))

# %% confusion matrix visualization
import seaborn as sns
f, ax = plt.subplots(figsize = (6,5))
sns.heatmap(cm_lr, annot = True, linewidths = 0.5, color = "red", fmt = ".0f", ax=ax)
plt.xlabel("y_predicted")
plt.ylabel("y_true")
plt.title("Confusion Matrix of Extreme Gradient Boosting")
plt.show()
```
